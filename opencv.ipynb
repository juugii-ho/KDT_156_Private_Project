```python
from datetime import datetime, timedelta
import datetime

datetime.datetime.today()
now = datetime.datetime.now()
nowDatetime = now.strftime('%Y-%m-%d %H:%M:%S')
nowDate = now.strftime('%Y-%m-%d')

default_font_path = '/Users/sml/Library/Fonts/SCDream6.otf'
asian_font_path = '/Users/sml/Library/Fonts/NotoSansKR-Bold.otf'

def contains_asian_characters(text):
    # 한자 및 일본어 문자 범위 확인
    for char in text:
        if '\u4e00' <= char <= '\u9fff' or '\u3040' <= char <= '\u30ff' or '\u3400' <= char <= '\u4dbf':
            return True
    return False
```

```python
# 각 텍스트별로 중앙 정렬 위치 계산
def draw_centered_text(draw, text, left, right, y, font_path, font_size, fill):
    font = ImageFont.truetype(font_path, font_size)
    text_width = font.getlength(text)
    x = left + ((right - left) - text_width) / 2
    draw.text((x, y), text, font=font, fill=fill)
```

```python
from PIL import Image
import os

def combine_images(image_paths, output_path):
    images = [Image.open(img) for img in image_paths]
    widths, heights = zip(*(i.size for i in images))

    total_height = sum(heights)
    max_width = max(widths)

    # 새 이미지 생성
    new_im = Image.new('RGB', (max_width, total_height))

    y_offset = 0
    for im in images:
        new_im.paste(im, (0, y_offset))
        y_offset += im.height

    new_im.save(output_path)

# 이미지 경로 설정
image_folder = 'rank'
image_files = [os.path.join(image_folder, f) for f in sorted(os.listdir(image_folder)) if f.endswith('.jpg')]

# 100개 이미지를 10개씩 나누어 합치기
for i in range(0, len(image_files), 10):
    batch = image_files[i:i + 10]
    output_filename = f'combined_{i//10 + 1:02}.jpg'  # 예: combined_1.jpg, combined_2.jpg, ...
    combine_images(batch, os.path.join(image_folder, output_filename))

# for i in range(0, len(image_files), 100):
#     batch = image_files[i:i + 100]
#     output_filename = f'total_combined_{i//100 + 1}.jpg'  # 예: combined_1.jpg, combined_2.jpg, ...
#     combine_images(batch, os.path.join(image_folder, output_filename))

print("이미지 합치기 완료!")
```

```python
# 새로운 가로 크기
new_width = 520

# 이미지 파일들의 경로
image_files = ['rank/combined_01.jpg', 'rank/combined_02.jpg', 'rank/combined_03.jpg', 'rank/combined_04.jpg', 'rank/combined_05.jpg', 'rank/combined_06.jpg', 'rank/combined_07.jpg', 'rank/combined_08.jpg', 'rank/combined_09.jpg', 'rank/combined_10.jpg']

# 최종 이미지 생성을 위한 리스트
images = []

for file in image_files:
    # 이미지 열기
    img = Image.open(file)
    
    # 이미지 크기 가져오기
    width, height = img.size
    
    # 새로운 세로 크기 계산 (가로 세로 비율 유지)
    new_height = int(height * (new_width / width))
    
    # 이미지 크기 조정
    resized_img = img.resize((new_width, new_height))
    
    # 리스트에 추가
    images.append(resized_img)

# 이미지들 세로로 합치기
total_height = sum(img.height for img in images)
new_img = Image.new('RGB', (new_width, total_height))

y_offset = 0
for img in images:
    new_img.paste(img, (0, y_offset))
    y_offset += img.height

# 결과 이미지 저장
new_img.save('rank/combined_final.jpg')
```
